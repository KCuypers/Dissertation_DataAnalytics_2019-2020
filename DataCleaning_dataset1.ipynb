```python
import pandas as pd

# dit is nodig om jupyter in widescreen te gebruiken
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
# Lees de dataset in en bekijk

filePath = r'C:\Case\DataCleaning\Ongevallen_Kennedy_2020_02_27_CSV.csv'
df = pd.read_csv(filePath, sep=';')

df.head(10)

df.shape
df.isnull().sum()
df.info()
df.apply(pd.Series.value_counts)
df.corr()

```

```python
# drop alle rijen die minder dan 18 niet-NaN waarden hebben
df = df.dropna(how='all',thresh=18)
df = df.reset_index(drop=True)

df.head(10)
```

```python
#drop alle lege kolommen
df = df.dropna(axis=1,how='all')

df.head(10)
```

```python
# Selecteer de kolommen die we later nodig hebben om tijdstip en locatie te bepalen

df_set1 = pd.DataFrame()

df_set1['set1_id'] = df['Ongeval ID']
df_set1['set1_datetime'] = df['Datum + uur']
df_set1['set1_ident8'] = df['Gewestweg (IDENT8)']
df_set1['set1_kmpunt'] = df['Km-punt (IDENT7)']

df_set1.shape
df_set1.head()
```

```python
#schrijf de resulterende tabel weg

df_set1.to_csv(r'C:\Case\DataCleaning\Dataset1_clean.csv',index=False)
```
